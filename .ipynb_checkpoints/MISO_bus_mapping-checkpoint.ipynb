This next cell uses a geopandas method (that basically just implements a SQL-style join) to assign all the lat/lon points given to us by MISO (so, solar/wind data locations) to retail service territories. Points may end up associated with multiple service territories if there are overlapping jurisdictions in the polygons.

```python
#join
match_gdf = gpd.sjoin(utilities_map, miso_gdf, op='contains')

#count
total_counts = len(miso_gdf.Name.unique())
matched_counts = len(match_gdf.Name.unique())

print('of '+str(total_counts)+' locations, '+str(matched_counts)+' receive utility matches')
```

This is pretty kludgy, but what I'm doing here is just pulling some code to map the points to a SEAMS zone via a linking from the retail service territories to which both the SEAMS zones and MISO points have now been matched. Again, it's possible a point is matched to more than one zone, which I've not really done cleaning for. Some points do not receive any match (you can see how many from the print at the end). I also assign a color from the dictionary for the mapping created previously; that's just for plotting.

```python
#create blank column for matches
match_gdf['SEAMS_ZONE'] = ['NA']*len(match_gdf.index)
match_gdf['SEAMS_COLOR'] = ['NA']*len(match_gdf.index)

#assign a SEAMS bus based on utility

for k,v in map_dict.items():
    for utility in v[0]:
        for i in match_gdf[match_gdf.NAME==utility].index:
            match_gdf.at[i,'SEAMS_ZONE']=k
            match_gdf.at[i,'SEAMS_COLOR']= v[1]

SEAMS_matched_counts = len(match_gdf[match_gdf.SEAMS_ZONE!='NA'].Name.unique())
print('of '+str(total_counts)+' locations, '+str(SEAMS_matched_counts)+' receive SEAMS matches')
```

I grabbed some functions online from the URL below for reformatting circles and rectangles in legends.

```python
# patch handlers
#https://stackoverflow.com/questions/40672088/matplotlib-customize-the-legend-to-show-squares-instead-of-rectangles

# --- handlers ---

class HandlerRect(HandlerPatch):

    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height,
                       fontsize, trans):

        x = width//2
        y = 0
        w = h = 10

        # create
        p = patches.Rectangle(xy=(x, y), width=w, height=h)

        # update with data from oryginal object
        self.update_prop(p, orig_handle, legend)

        # move xy to legend
        p.set_transform(trans)

        return [p]


class HandlerCircle(HandlerPatch):

    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height,
                       fontsize, trans):

        r = 5
        x = r + width//2
        y = height//2

        # create 
        p = patches.Circle(xy=(x, y), radius=r)

        # update with data from oryginal object
        self.update_prop(p, orig_handle, legend)

        # move xy to legend
        p.set_transform(trans)

        return [p]
```

```python
#do some plotting on map of USA to check that things look reasonable
#color matched points by their assigned SEAMS zone

fig, ax = plt.subplots(1, figsize=(8, 8))
myaxes = plt.axes()
myaxes.set_ylim([23, 50])
myaxes.set_xlim([-108,-82])
myaxes.set_title('(Attempted) Bus Mapping to SEAMS zones')
iso_map[iso_map['NAME']==iso_map.at[0,'NAME']].plot(ax=myaxes,facecolor='b',edgecolor='r',alpha=0.1,linewidth=2)
states.plot(ax=myaxes,edgecolor='k',facecolor='None')


#hopefully this dataframe will have more matched zones, so then it'll plot more
match_gdf[match_gdf.SEAMS_ZONE!='NA'][['Lat','Lon','SEAMS_ZONE','SEAMS_COLOR']].plot.scatter(ax=myaxes,x='Lon',
                      y='Lat',c=colormap[categories])
#this dataframe should eventually be empty, and can then be eliminated, since every point should be matched
unmatched_points.plot.scatter(ax=myaxes,x='Lon',y='Lat',c='k')

leg_labels = []
for k,v in map_dict.items():
    leg_labels.append(patches.Circle((0,0),1,color=v[1],label=k))
leg_labels.append(patches.Circle((0,0),1,color='k',label='Unmatched'))
leg_labels.append(patches.Rectangle((0,0),1,1,facecolor='b',edgecolor='r',alpha=0.1,label='MISO'))
myaxes.legend(handles=leg_labels,handler_map={
               patches.Rectangle: HandlerRect(),
               patches.Circle: HandlerCircle(),
            },fontsize=11,loc='lower left')

#plt.show() #uncomment this if you'd rather just show the plot inline
#this just writes the plot to a folder of your choice
os.chdir(r'C:\Users\llavin\Desktop\PRAS_RA_MISO_930')
plt.savefig("miso_locs_to_SEAMS_zones.jpg", dpi=300)
```
